```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.shape
```

OBSERVATION: There are 150 datapoints that is 150 flowers and has 4 features and one class label.

```python
iris.columns
```

```python
labels = iris['species']
data = iris.drop('species',axis=1)
```

OBSERVATION: These are the feature names of the data.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3,random_state=12)
pca_data = pca.fit_transform(data)
transformed_data = pd.DataFrame(pca_data,columns=['pcomponent1','pcomponent2','pcomponent3'])
transformed_data = pd.concat([transformed_data,labels],axis=1)
transformed_data.head()
```

```python
x = transformed_data['pcomponent1']
y = transformed_data['pcomponent2']
z = transformed_data['pcomponent3']
```

```python
#plotting a 3d-chart
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(7,7))
ax = plt.axes(projection='3d')

ax.grid(b=True,color = 'grey',linestyle ='-.', linewidth = 0.3, 
        alpha = 0.2)
c_map = plt.get_cmap('hsv')
sct = ax.scatter3D(x,y,z,alpha=0.8,c=(x+y+z),cmap = c_map,marker='^')

plt.title("iris 3D scatter plot")
ax.set_xlabel('X-axis') 
ax.set_ylabel('Y-axis') 
ax.set_zlabel('Z-axis')

plt.show()
```
